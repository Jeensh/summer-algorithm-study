```python
# 스택 - 10828번
# sol1 - 시간 초과
class stack:
    item = []
    
    def push(self, item):
        self.item.append(item)
    def pop(self):
        index = len(self.item)-1
        if index < 0:
            print(-1)
        else:
            temp = self.item[index]
            del self.item[index]
            print(temp)
    def size(self):
        print(len(self.item))
    def empty(self):
        if len(self.item) == 0:
            print(1)
        else:
            print(0)
    def top(self):
        index = len(self.item)-1
        if index < 0:
            print(-1)
        else:
            print(self.item[index])
            
if __name__ == "__main__":
    n = int(input()) # 명령 개수 입력
    stack = stack()
    
    for i in range(n):
        order = input()
        if 'push' in order:
            order,item = order.split()
            stack.push(item)
        else:
            if order == 'pop':
                stack.pop()
            elif order == 'size':
                stack.size()
            elif order == 'empty':
                stack.empty()
            elif order == 'top':
                stack.top()
```

```python
# sol2 - 시간초과
if __name__ == "__main__":
    n = int(input()) # 명령 개수 입력
    items = []
    
    for i in range(n):
        order = input()
        if 'push' in order:
            order,item = order.split()
            items.append(item)
        else:
            index = len(items) - 1
            if order == 'pop':
                if index < 0:
                    print(-1)
                else :
                    print(items[index])
                    del items[index]
            elif order == 'size':
                print(index + 1)
            elif order == 'empty':
                if index < 0:
                    print(1)
                else:
                    print(0)
            elif order == 'top':
                if index < 0:
                    print(-1)
                else :
                    print(items[index])
```

```python
# sol3 - 정답처리됨
import sys

if __name__ == "__main__":
    n = int(sys.stdin.readline().strip()) # 명령 개수 입력
    items = []
    
    for i in range(n):
        order = sys.stdin.readline().strip()
        if 'push' in order:
            order,item = order.split()
            items.append(item)
        else:
            index = len(items) - 1
            if order == 'pop':
                if index < 0:
                    print(-1)
                else :
                    print(items[index])
                    del items[index]
            elif order == 'size':
                print(index + 1)
            elif order == 'empty':
                if index < 0:
                    print(1)
                else:
                    print(0)
            elif order == 'top':
                if index < 0:
                    print(-1)
                else :
                    print(items[index])
```

```python
# 단어 뒤집기 - 9093번
import sys

n = int(sys.stdin.readline().strip())

for i in range(n):
    s = sys.stdin.readline().strip().split() # 입력받은 문장을 단어 단위로 잘라서 리스트에 삽입
    for word in s:
        print(word[::-1],end=' ')
    print()
```

```python
# 괄호 - 9012번
import sys

n = int(sys.stdin.readline().strip())
stack = []

for i in range(n):
    ps = sys.stdin.readline().strip()
    for c in ps:
        if c == '(':        # 문자가 '('이면 스택에 저장
            stack.append(c)
        else:               # 문자가 ')'일 경우
            if len(stack) == 0: # 스택에 저장된 문자가 없으면 'NO'
                check = ')'
                break
            else:               # 스택에 저장된 문자가 있을 경우 추출
                check = stack.pop()
                if check == '(':    # 추출한 문자가 '('일 경우 다음 문자로 진행
                    continue
    if len(stack) != 0 or check == ')': # VPS인지 검증
        print('NO')
    else:
        print('YES')
    stack.clear()
```

```python
# 스택 수열 - 1874번
import sys

n = int(sys.stdin.readline().strip())
l = []
stack = []
sequence = []
temp = []

for i in range(n): # 수열 입력받기
    l.append(int(sys.stdin.readline().strip()))

sorted_l = sorted(l) # 입력 받은 수열 오름차순으로 정리

count = 0

for i in l: # 입력 받은 수열의 순서대로 진행
    while count <= len(sorted_l)-1 or len(stack) != 0: # 아직 push하지 않은 수가 남았거나 stack이 empty가 아니면
        if count < len(sorted_l) and sorted_l[count] < i: # 현재 참조하는 수열의 수 보다 push하지 않은 수가 작을 때
            stack.append(sorted_l[count])       # 스택에 push
            sequence.append('+')
            count += 1
        elif len(stack) > 0 and stack[len(stack)-1] == i: # 스택의 top이 현재 참조하는 수열의 수와 같을 때
            sequence.append('-')                # 스택을 pop
            temp.append(stack.pop())
            break
        elif count < len(sorted_l) and sorted_l[count] == i: # 현재 참조하는 수열의 수와 push하지 않는 수가 같을 때
            sequence.append('+')                # 스택을 push
            temp.append(sorted_l[count])
            count += 1
            sequence.append('-')                # 스택을 pop
            break
        else:
            break

if temp == l:
    for i in sequence:
        print(i)
else:
    print('NO')
```

```python
# 에디터 - 1406번
# sol1 - 시간초과
import sys

s = list(sys.stdin.readline().strip())
n = int(sys.stdin.readline().strip())

pos = len(s)
for i in range(n):
    order = sys.stdin.readline().strip()
    if 'P' in order:
        order, item = order.split()
        s.insert(pos,item)
        pos += 1
    elif order == 'L':
        if pos > 0:
            pos -= 1
    elif order == 'D':
        if pos < len(s):
            pos += 1
    elif order == 'B':
        if pos > 0:
            del s[pos-1]
            pos -= 1

for i in s:
    print(i,end='')
```

```python
# sol2 - 시간초과
import sys

s = sys.stdin.readline().strip()
n = int(sys.stdin.readline().strip())

pos = len(s)
for i in range(n):
    order = sys.stdin.readline().strip()
    if 'P' in order:
        order, item = order.split()
        s = s[:pos] + item + s[pos:] 
        pos += 1
    elif order == 'L':
        if pos > 0:
            pos -= 1
    elif order == 'D':
        if pos < len(s):
            pos += 1
    elif order == 'B':
        if pos > 0:
            s = s[:pos-1] + s[pos:]
            pos -= 1
print(s)
```

```python
# sol3 - 다른 사람 풀이 참고(스택 2개 이용)
import sys

left = list(sys.stdin.readline().strip())
right = []
n = int(sys.stdin.readline().strip())

for i in range(n):
    order = sys.stdin.readline().strip()
    if order[0] == 'P':
        left.append(order[2])
    elif order[0] == 'L':
        if left:
            right.append(left.pop())
    elif order[0] == 'D':
        if right:
            left.append(right.pop())
    elif order[0] == 'B':
        if left:
            left.pop()
            
print(''.join(left[:]+right[::-1]))
```

```python
# 큐 - 10845번
import sys

n = int(sys.stdin.readline().strip())
queue = []

for i in range(n):
    order = sys.stdin.readline().strip()
    if 'push' in order:
        order,item = order.split()
        queue.append(int(item))
    elif order == 'pop':
        if queue:
            item = queue[0]
            del queue[0]
            print(item)
        else:
            print(-1)
    elif order == 'size':
        print(len(queue))
    elif order == 'front':
        if queue:
            item = queue[0]
            print(item)
        else:
            print(-1)
    elif order == 'empty':
        if queue:
            print(0)
        else:
            print(1)
    elif order == 'back':
        if queue:
            item = queue[len(queue)-1]
            print(item)
        else:
            print(-1)
```

```python
# 조세퍼스 문제 - 1158번
import sys

n,k = map(int,sys.stdin.readline().strip().split())
l = []
sequence = []
index = 0

for i in range(n):
    l.append(i+1)

while l:
    index = (index + k) % len(l) -1
    if index < 0:
        index += len(l)
    sequence.append(l[index])
    del l[index]

print('<' + str(sequence[0]),end='')
for i in sequence[1:]:
    print(', ' + str(i),end='')
print('>')
```

```python
# 덱 - 10866번port sys

n = int(sys.stdin.readline().strip())
deque = []

for i in range(n):
    order = sys.stdin.readline().strip()

    if 'push_front' in order:
        order, item = order.split()
        deque.insert(0,item)
    elif 'push_back' in order:
        order, item = order.split()
        deque.append(item)
    elif order == 'pop_front':
        if deque:
            item = deque[0]
            del deque[0]
            print(item)
        else:
            print(-1)
    elif order == 'pop_back':
        if deque:
            print(deque.pop())
        else:
            print(-1)
    elif order == 'size':
        print(len(deque))
    elif order == 'empty':
        if deque:
            print(0)
        else:
            print(1)
    elif order == 'front':
        if deque:
            item = deque[0]
            print(item)
        else:
            print(-1)
    elif order == 'back':
        if deque:
            print(deque[len(deque)-1])
        else:
            print(-1)
```
