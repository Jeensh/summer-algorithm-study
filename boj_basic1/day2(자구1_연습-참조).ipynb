### 미해결 문제

- 오큰수 : 시간초과
- 오등큰수 : 시간초과

```python
# 단어 뒤집기2 - 17413번
import sys

s = sys.stdin.readline().strip()
stack = []
result = []
switch = 0

for c in s:
    if c == '>':
        result.append(c)
        switch = 0
    elif switch:
        result.append(c)
    elif c == '<':
        if stack:
            for i in range(len(stack)):
                result.append(stack.pop())
        result.append(c)
        switch = 1
    elif c == ' ':
        if stack:
            for i in range(len(stack)):
                result.append(stack.pop())
        result.append(c)
    else:
        stack.append(c)
        
if stack :
    for i in range(len(stack)):
        result.append(stack.pop())

print(''.join(result))
```

```python
# 쇠막대기
# sol1 = 시간초과
import sys

s = sys.stdin.readline().strip()
on = []
result = 0

j = s[0]
for i in s[1:]:
    if j == '(':
        if i == ')': 
            for k in range(len(on)):
                on[k] += 1
        else:
            on.append(0)
    else:
        if i == ')':
            result += (on.pop() + 1)
    j = i
print(result)
```

```python
# sol2 - 시간초과
import sys

s = sys.stdin.readline().strip()
on = []
result = 0

for i in range(1,len(s)):
    if s[i-1] == '(':
        if s[i] == ')': 
            for k in range(len(on)):
                on[k] += 1
        else:
            on.append(0)
    else:
        if s[i] == ')':
            result += (on.pop() + 1)
print(result)
```

```python
# sol3 다른 사람 코드 참조
bar_razor = list(input())
answer = 0
stack = []

for i in range(len(bar_razor)):
    if bar_razor[i] == '(': #스택 쌓기
        stack.append('(')
        
    else:
        if bar_razor[i-1] == '(': #()라면 (를 pop하고 현재 스택에 들어있는 ( 수만큼 값을 더해준다.
            stack.pop()
            answer += len(stack)
            
        else:
            stack.pop() 
            answer += 1 #끄트머리 막대기 부분을 더해준다

print(answer)
```

```python
# 오큰수 - 17298번 - 시간초과
# sol1
import sys

n = int(sys.stdin.readline().strip())
series = list(map(int,sys.stdin.readline().strip().split()))
result = []

for i in range(n):
    
    for j in range(i+1,n):
        if series[i] < series[j]:
            result.append(series[j])
            break
        if j == n-1:
            result.append(-1)

for i in result:
    print(i,end=' ')
print('-1')
```

```python
# sol2 - 다른 사람 코드 참고
numbers = int(input())
num_list = list(map(int, input().split()))
stack = []
result = [-1 for _ in range(numbers)]

for i in range(len(num_list)):
    try:
        while num_list[stack[-1]] < num_list[i]:
            result[stack.pop()] = num_list[i]
    except:
        pass
        
    stack.append(i)
        
for i in range(numbers):
    print(result[i], end = ' ')
```

```python
# 오등큰수 - 17299번
# sol1 - 시간초과
import sys

n = int(sys.stdin.readline().strip())
series = list(map(int,sys.stdin.readline().strip().split()))
result = [-1 for i in range(n)]
stack =[]
count = [0 for i in range(n)]

for i in range(n):
    count[i] = series.count(series[i])

for i in range(n):
    if stack:
        if count[stack[-1]] < count[i]:
            while count[stack[-1]] < count[i]:
                result[stack.pop()] = series[i]
            stack.append(i)
        else:
            stack.append(i)
    else:
        stack.append(i)

for i in result:
    print(i,end=' ')
```

```python
# sol2 - 다른 사람 코드 참조
import sys
input = sys.stdin.readline

from collections import Counter

n = int(input())
aa = list(map(int,input().split()))
ff = Counter(aa)
stack = []
ngf = [-1] * n

for i in range(n):
    while stack and ff[aa[stack[-1]]] < ff[aa[i]]:
        ngf[stack.pop()] = aa[i]
    stack.append(i)

print(*ngf)
```

```python
# 후위 표기식2
import sys

n = int(sys.stdin.readline().strip())
formula = sys.stdin.readline().strip()
operand = []
stack = []
for i in range(n):
    operand.append(int(sys.stdin.readline().strip()))

count = 0
for i in formula:
    if ord(i) < 65 or ord(i) > 90:
        b = stack.pop()
        a = stack.pop()
        if i == '+':
            stack.append(a + b)
        elif i == '-':
            stack.append(a - b)
        elif i == '*':
            stack.append(a * b)
        elif i == '/':
            stack.append(a / b)
    else:
        stack.append(operand[ord(i)-65])

print(format(stack.pop(),".2f"))
```

```python
# 후위 표기식 - 풀이 실패 - 다른 사람 코드 참고
import sys
input = sys.stdin.readline
 
prior = {"/": 2, "*": 2, "+": 1, "-": 1, "(": 0}
stack = []
_str = input().rstrip()
 
for s in _str:
    if s.isalpha():
        print(s, end="")
    elif s == "(":
        stack.append(s)
    elif s == ")":
        while True:
            temp = stack.pop()
            if temp == "(":
                break
            print(temp, end="")
    else:
        while stack and prior[stack[-1]] >= prior[s]:
            print(stack.pop(), end="")
        stack.append(s)
        
while stack:
    print(stack.pop(), end="")
```

```python
# 알파벳 찾기
import sys

s = sys.stdin.readline().strip()
alpha = [-1 for i in range(26)]

count = 0
for c in s:
    if alpha[ord(c)-97] == -1:
        alpha[ord(c)-97] = count
    count += 1
    
print(*alpha)
```

```python
# 문자열 분석
import sys

for i in range(100):
    s = sys.stdin.readline()
    if not s:
        break
    result = [0 for i in range(4)]

    for c in s:
        if ord(c) >= ord('a') and ord(c) <= ord('z'):
            result[0] += 1
        elif ord(c) >= ord('A') and ord(c) <= ord('Z'):
            result[1] += 1
        elif ord(c) >= ord('0') and ord(c) <= ord('9'):
            result[2] += 1            
        elif c == ' ':
            result[3] += 1

    print(*result)
```

```python
# ROT13
import sys

s = sys.stdin.readline()
result = ''
for c in s:
    if ord(c) >= ord('a') and ord(c) <= ord('z'):
        result += chr((ord(c) - ord('a')+13)%26 + ord('a'))
    elif ord(c) >= ord('A') and ord(c) <= ord('Z'):
        result += chr((ord(c) - ord('A')+13)%26 + ord('A'))
    else:
        result += c

print(result)
```
