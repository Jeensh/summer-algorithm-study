# 백준 기초 -1 (자료구조)
### 미해결 문제
- -2 진수 : 다른사람 코드 참고
- 골드바흐 파티션 : 시간초과

```python
# GCD 합
import sys

def get_GCD(a,b):
    if a < b:
        temp = a
        a = b
        b = temp
        
    r = -1
    while r != 0:
        r = a % b
        a = b
        b = r
    return a
        

n = int(sys.stdin.readline().strip())

for i in range(n):
    l = list(map(int,sys.stdin.readline().strip().split()))
    sum = 0
    for i in range(1,l[0]+1):
        for j in range(i+1,l[0]+1):
            sum += get_GCD(l[i],l[j])
    print(sum)
```

```python
# 숨바꼭질6
import sys

def get_GCD(a,b):
    if a < b:
        temp = a
        a = b
        b = temp
        
    r = -1
    while r != 0:
        r = a % b
        a = b
        b = r
    return a

n,s = map(int,sys.stdin.readline().strip().split())
l = list(map(int,sys.stdin.readline().strip().split()))

for i in range(n):
    l[i] = abs(s - l[i])

for i in range(n-1):
    l[i+1] = get_GCD(l[i],l[i+1])
print(l[-1])
```

```python
# - 2진수 - 다른 사람 코드 참고
import sys

n = int(sys.stdin.readline().strip())

a = n
b = a//-2
r = a%-2
if r == -1:
    b += 1
    r = 1
result = []
result.append(r)
while b != 0:
    a = b
    b = a//-2
    r = a%-2
    if r == -1:
        b += 1
        r = 1
    result.append(r)

for i in result[::-1]:
    print(i,end='')
```

```python
# 골드바흐 파티션
# sol1 - 시간초과
import sys

n = int(sys.stdin.readline().strip())

m = [1 for i in range(1000000)]
m[0] = 0
for i in range(2,1001):
    if m[i-1]:
        for j in range(i+i,1000001, i):
            m[j-1] = 0

for k in range(n):
    t = int(sys.stdin.readline().strip())
    count = 0
    for i in range(t):
        if m[i]:
            for j in range(i,t):
                if m[j]:
                    if i + j + 2 == t:
                        count += 1
                        break
                elif i + j + 2 > t:
                    break
    print(count)
```

```python
# sol2 - 다른 사람 코드 참조
prime_arr = []
sieve = [False, False] + ([True] * (1000000 - 1))
for i in range(2, 1000001):
    if sieve[i]:
        prime_arr.append(i)
    for j in range(i * 2, 1000001, i):
        sieve[j] = False

ans = []
for tc in range(int(input())):

    N = int(input())
    cnt = 0
    for i in range(len(prime_arr)):
        if prime_arr[i] > N - prime_arr[i]:
            break
        if sieve[N - prime_arr[i]]:
            cnt += 1
    ans.append(cnt)

print(*ans, sep='\n')
```

```python
# 진법 변환2
import sys

n,t = map(int,sys.stdin.readline().strip().split())

l = [chr(i) for i in range(65,91)]
result = []

a = n
b = a//t
r = a%t
if r >= 10:
    r = l[r-10]
result.append(r)
while b != 0:
    a = b
    b = a//t
    r = a%t
    if r >= 10:
        r = l[r-10]
    result.append(r)
for i in result[::-1]:
    print(i,end='')
```

```python
# 진법 변환
import sys

n,t = sys.stdin.readline().strip().split()
sum = 0

for i in range(len(n)-1,-1,-1):
    if  ord(n[i]) >= 65 and ord(n[i]) <= 90:
        sum += (ord(n[i])-55) * (int(t)**(len(n)-i-1))
    else:
        sum += int(n[i]) * (int(t)**(len(n)-i-1))
print(sum)
        
```

```python
# Base Conversion
import sys

a,t = map(int,sys.stdin.readline().strip().split())
count = int(sys.stdin.readline().strip())
n = list(map(int,sys.stdin.readline().strip().split()))

sum = 0
for i in range(1,count+1):
    sum += n[count-i] * a**(i-1)

result = []
a = sum
b = a//t
r = a%t
result.append(r)
while b != 0:
    a = b
    b = a//t
    r = a%t
    result.append(r)
    
print(*result[::-1])
```
